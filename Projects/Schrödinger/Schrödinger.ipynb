```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
#Definiciones iniciales

#Parte espacial
#El índice espacial es jota, el cuál va de 0 a N
h=0.1
N=500

#Parte temporal
#El índices temporal es n, el cuál va de 0 a t 
s=0.1
t=1000

#Número de ciclos (entre 1 y N/4)
N_ciclos=50

#Lambda
lamb=0.4
```

```python
#Definimos variables iniciales
ko=(2*np.pi*N_ciclos)/N

s=1/(4*ko**2)


#Potencial
V=np.zeros(N+1)
for jota in range(N+1):
    if (jota>=(2*N)/5) and (jota<=(3*N)/5):
        V[jota]=lamb*ko**2
    else:
        V[jota]=0

        
#Función de onda inicialmente        
phi=np.zeros((N+1,t), dtype=complex)
for jota in range(N+1):
    phi[jota][0]=np.exp((ko*jota)*1j)*np.exp((-8*((4*jota)-N)**2)/N**2)


    
print(ko)
print()

print(s)
print()

print(V)
print()

print(phi)
print()

#Normal inicialmente
norma = np.linalg.norm(phi, axis=0)
print(norma)
```

```python
#Definimos alpha
alpha=np.zeros(N, dtype=complex)
A=1

for jota in range(N-1, -1, -1):
    A0=-2+(2j/s)-V[jota]
    alpha[jota-1]=-A/(A0+A*alpha[jota])
    

print()
print(alpha)
```

```python
#Definimos beta
beta=np.zeros((N,t), dtype=complex)

#Definimos chi
chi=np.zeros((N+1,t), dtype=complex)


#Empezamos el bucle que acompaña al tiempo
for n in range(t-1):
    
    for jota in range(N-1, -1, -1):
        b=(4j*phi[jota][n])/s
        A0=-2+(2j/s)-V[jota]
        beta[jota-1][n]=(b-A*beta[jota][n])/(A0+A*alpha[jota])
        
    for jota in range(N-1):
        chi[jota+1][n]=alpha[jota]*chi[jota][n]+beta[jota][n]   

    for jota in range(N+1):
        phi[jota][n+1]=chi[jota][n]-phi[jota][n]
    
    
print(beta)
print()

print(chi)
print()

print(phi)
print()

```

```python
norma=np.zeros((N,t))

#calculo
for n in range(t):
    suma=0
    for jota in range(N):
        suma=suma+((((abs(phi[jota,n]))**2+(abs(phi[jota+1,n]))**2)/2)*h)
        norma[jota][n]=suma
        
print(norma)  
```
