```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

# Gráficas

```python
x=2*np.pi*np.linspace(0,1,100)  #Linspace devuelve números espaciados uniformemente en un intervalo específico
y=np.sin(x)    #Variable y

plt.figure(figsize=(6,5), layout='tight')   #Ajustar tamaño gráfica

plt.title('Mi primer gráfica')   #Título gráfico
plt.plot(x,y,'r', label='sin(x)')        #Graficar, r es red, rojo
plt.xlabel('Ángulo en radianes')     #Etiqueta eje x
plt.ylabel('sin(x)')         #Etiqueta eje y
plt.legend()                 #Leyenda

plt.axis([0, 2*np.pi, -1, 1])   #Establece los límites del eje x e y, respectivamente
plt.show()                       #Mostrar gráfico
```

```python
x=2*np.pi*np.linspace(0,1,100)    #Linspace devuelve números espaciados uniformemente en un intervalo específico
y=np.sin(x)    #Variable y

plt.plot(x,y, label=r'$\sin(\theta)$')        #Graficar
plt.xlabel(r'$\theta$', fontsize=16)     #Etiqueta eje x
plt.ylabel(r'$\sin(\theta)$', fontsize=16)         #Etiqueta eje y
plt.legend()                 #Leyenda

plt.title(r'Mi primer gráfica') #Título gráfico
plt.axis([0, np.pi, -1, 1])   #Establece los límites del eje x e y, respectivamente
plt.show()                       #Mostrar gráfico
```

### Graficar dos funciones 

```python
x=2*np.pi*np.linspace(0,1,100)  #Variable x
y1=np.cos(x)  #Variable y1
y2=np.sin(x)  #Variable y2
 
plt.figure(figsize=(6,4), layout='tight')   #Ajustar tamaño gráfica
plt.plot(x,y1, 'r', label=r'$\cos(\theta)$')        #Graficar, con 'r' cambia el color y el estilo 
plt.plot(x,y2, 'b', label=r'$\sin(\theta)$')        #de la línea
plt.xlabel(r'$\theta$', fontsize=16)     #Etiqueta eje x
plt.ylabel(r'$\sin(\theta)$', fontsize=16)         #Etiqueta eje y
plt.legend()                 #Leyenda

plt.title(r'Dos funciones')  #Título gráfico
plt.axis([0, 2*np.pi,-1,1])  #Límites del gráfico
plt.show()                       #Mostrar gráfico
```

### Subgráficos

```python
fig, axs=plt.subplots(ncols=2, nrows=2, figsize=(3.5,2.5), layout="constrained")  
#Añadimos una etiqueta
for row in range(2):
    for col in range(2):
        axs[row,col].annotate(f'axs[{row},{col}]', (0.5,0.5),
                             transform=axs[row, col].transAxes,
                             ha='center', va='center',fontsize=18,
                             color='darkgrey')
        
```

### Tarea

```python
fig, axs=plt.subplots(2 , figsize=(3.5,3.5), layout="constrained") # El 2 es el número de gráficos
x=2*np.pi*np.linspace(0,1,100)
y1=np.cos(x)
y2=np.sin(x)

axs[0].plot(x,y1, 'r',label=r'$\cos(\theta)$') #Para graficar el gráfico 1
axs[0].set_xlabel(r'$\theta$', fontsize=16)     #Etiqueta eje x
axs[0].set_ylabel(r'$\sin\theta$', fontsize=16)         #Etiqueta eje y
axs[0].legend()                 #Leyenda
axs[0].set_title(r'Primera')

axs[1].plot(x,y2, 'b', label=r'$\sin(\theta)$')    #Para graficar el gráfico 2   
axs[1].set_xlabel(r'$\theta$', fontsize=16)     #Etiqueta eje x
axs[1].set_ylabel(r'$\cos\theta$', fontsize=16)         #Etiqueta eje y
axs[1].legend()     
axs[1].set_title(r'Segunda')

fig.suptitle('Titulo')


                           
        
```

```python
fig, axs=plt.subplots(ncols=2,nrows=2, figsize=(7,5), layout="constrained")
x=2*np.pi*np.linspace(0,1,100)
y1=np.cos(x)
y2=np.sin(x)

axs[0,0].plot(x,y1, 'r',label=r'$\cos(\theta)$')
axs[0,0].set_xlabel(r'$\theta$', fontsize=16)     #Etiqueta eje x
axs[0,0].set_ylabel(r'$\sin\theta$', fontsize=16)         #Etiqueta eje y                #Leyenda
axs[0,0].set_title(r'Primera')

axs[1,1].plot(x,y2, 'b', label=r'$\sin(\theta)$')       
axs[1,1].set_xlabel(r'$\theta$', fontsize=16)     #Etiqueta eje x
axs[1,1].set_ylabel(r'$\cos\theta$', fontsize=16)         #Etiqueta eje y   
axs[1,1].set_title(r'Segunda')

axs[0,1].axis('off')
axs[1,0].axis('off')

fig.suptitle('Titulo')
```

### Hacer anotaciones en la gráfica

```python
x=2*np.pi*np.linspace(0,1,100)
y1=np.cos(x)
y2=np.sin(x)
fig, ax=plt.subplots()


ax.plot(x,y1, 'r', label=r'$\cos(\theta)$')        #Graficar, con 'r' cambia el color y el estilo 
ax.plot(x,y2, 'b', label=r'$\sin(\theta)$')        #de la línea
ax.set_xlabel(r'$\theta$', fontsize=16)     #Etiqueta eje x
ax.set_ylabel(r'$\sin(\theta)$', fontsize=16)         #Etiqueta eje y
ax.legend()
ax.set_xticks([0.,0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi], labels=["0",r"$\frac{\pi}{2}$", r"$\pi$"\
                                                                 ,r"$\frac{3\pi}{2}$", r"$2\pi$"]) #Poner marcas en el eje x
ax.set_yticks([-1,0.,1])   #Poner marcas en el eje y

ax.annotate('Máximo',xy=(1.61,1), xytext=(2.8,1.02), 
           arrowprops=dict(facecolor='black', width=2, headwidth=5,
                           headlength=5, shrink=0.05))   #hacer anotaciones
ax.annotate('Intersección',xy=(3.92,-0.7), xytext=(3.4,-1.08), 
           arrowprops=dict(facecolor='black', width=2, headwidth=5,
                           headlength=5, shrink=0.05))   #hacer anotaciones
ax.annotate('Intersección eje x',xy=(1.61,0), xytext=(-0.3,-0.5),
           arrowprops=dict(facecolor='black', width=2, headwidth=5,
                           headlength=5, shrink=0.05))   #hacer anotaciones
ax.text(3.5, -0.6, "Intersección", bbox=dict(boxstyle='ellipse, pad=0.4', fc='none'))

plt.show()
```

### Hacer zoom

```python
x = 2*np.pi*np.linspace(0,1,100)
y1 = np.cos(x)
y2 = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x,y2, 'b', label = r'$\sin(\theta)$')
ax.set_xlabel(r'$\theta$', fontsize = 16)
ax.set_ylabel(r'$f(\theta)$', fontsize = 16)
ax.set_xticks([0., 0.5*np.pi, np.pi, 1.5*np.pi, 2*np.pi], labels = ["$0$", r"$\frac{1}{2}\pi$",
                       r"$\pi$", r"$\frac{3}{2}\pi$", r"$2\pi$"])
ax.set_yticks([-1.0, 0., 1])

x_zoom = 2*np.pi*np.linspace(0.2,0.3,100)
y_zoom = np.sin(x_zoom)

axins = ax.inset_axes(
    [0.2, 0.35, 0.3, 0.3])
axins.plot(x_zoom, y_zoom, 'r')
axins.set_xticks([x_zoom[0], x_zoom[50], x_zoom[99]], 
                 labels=[r"$\frac{4\pi}{10}$",r"$\frac{5\pi}{10}$",r"$\frac{6\pi}{10}$"])
ax.indicate_inset_zoom(axins, edgecolor="black")
plt.show()
```
